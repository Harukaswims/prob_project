```python
#splitting the dataset
x = df.drop('prognosis', axis = 1)
y = df['prognosis']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
#decision tree classifier
tree = DecisionTreeClassifier(random_state=42)
tree.fit(x_train, y_train)
tree_pred=tree.predict(x_test)
tree_acc=accuracy_score(y_test, tree_pred)
print("Decision Tree Classifier:")
print("Accuracy on test set: {:.2f}%".format(tree_acc * 100))
```

```python
from sklearn.tree import plot_tree
col=df.columns
plt.figure(figsize=(90,50))
plot_tree(tree,filled='True',feature_names=col, class_names=[str(i) for i in tree.classes_])
plt.show()
```

```python
forest=RandomForestClassifier(random_state=42)
fr=forest.fit(x_train,y_train)
prediction=forest.predict(x_test)
prediction
```

```python
acc=accuracy_score(y_test,prediction)
print("Random Forest Classifier:")
print("Accuracy on test set: {:.2f}%".format(acc * 100))
```

```python
confusion=confusion_matrix(y_test,prediction)
ConfusionMatrixDisplay(confusion_matrix=confusion).plot()
```

```python
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
logreg_pred = logreg.predict(x_test)
logreg_acc = accuracy_score(y_test, logreg_pred)
logreg_confidence = logreg.predict_proba(x_test)
print("Logistic Regression Classifier:")
print("Accuracy on test set: {:.2f}%".format(logreg_acc * 100))
```

```python
def evaluate_classifier(classifier, x_train, x_test, y_train, y_test):
    start_time = time.time()
    classifier.fit(x_train, y_train)
    train_time = time.time() - start_time

    start_time = time.time()
    y_pred = classifier.predict(x_test)
    test_time = time.time() - start_time

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    return accuracy, precision, recall, f1, train_time, test_time
```

```python
accuracy_a,precision_a,recall_a,f1_a,train_time_a,test_time_a = evaluate_classifier(logreg, x_train,x_test,y_train,y_test)
accuracy_b,precision_b,recall_b,f1_b,train_time_b,test_time_b = evaluate_classifier(tree, x_train, x_test,y_train,y_test)
accuracy_c,precision_c,recall_c,f1_c,train_time_c,test_time_c = evaluate_classifier(forest, x_train,x_test,y_train,y_test)
```

```python
print("Logistic Regression - Accuracy: {:.2f}, Precision: {:.2f}, Recall: {:.2f}, F1 Score: {:.2f}, Training Time: {:.4f}s, Testing Time: {:.4f}s".format(accuracy_a, precision_a, recall_a, f1_a, train_time_a, test_time_a))
print("Decision Tree - Accuracy: {:.2f}, Precision: {:.2f}, Recall: {:.2f}, F1 Score: {:.2f}, Training Time: {:.4f}s, Testing Time: {:.4f}s".format(accuracy_b, precision_b, recall_b, f1_b, train_time_b, test_time_b))
print("Random Forest - Accuracy: {:.2f}, Precision: {:.2f}, Recall: {:.2f}, F1 Score: {:.2f}, Training Time: {:.4f}s, Testing Time: {:.4f}s".format(accuracy_c, precision_c, recall_c, f1_c, train_time_c, test_time_c))
```
